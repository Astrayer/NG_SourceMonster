```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib notebook
```

```python
# Import necessary Data
df = pd.read_pickle('/home/alan/Desktop/Civil_Engineering/Thesis_Research/Pickle_Jar/Generation_and_FuelData.pkl')
plants = pickle.load(open('/home/alan/Desktop/Civil_Engineering/Thesis_Research/Pickle_Jar/PlantIDs_NG.pkl', 'rb'))
```

```python
all_deets = df.loc[df['Plant Id'].isin(plants)]
effish = all_deets[['Plant Name',
                    'Plant State',
                   'Reported\nPrime Mover',
                   'Elec Fuel Consumption\nMMBtu',
                   'Net Generation\n(Megawatthours)']]
effish
```

```python
effish = effish.apply(lambda x:x*1e3 if x.name in ['Net Generation\n(Megawatthours)'] else x)
```

```python
effish = effish.apply(lambda x:x*1e6 if x.name in ['Elec Fuel Consumption\nMMBtu'] else x)
```

```python
effish['Heat Rate (Btu/kWh)'] = (effish['Elec Fuel Consumption\nMMBtu']/effish['Net Generation\n(Megawatthours)'])
```
