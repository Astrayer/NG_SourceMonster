```python
df = pd.read_pickle('../pickleJar/2__Plant_Y2017.pkl')
plant = pickle.load(open('../pickleJar/PlantIDs_NG.pkl', 'rb'))
```

```python
df = df[['Plant Name',
         'Plant Code',
         'Sector Name',
         'State',
         'Natural Gas LDC Name', 
         'Natural Gas Pipeline Name 1',
         'Natural Gas Pipeline Name 2',
         'Natural Gas Pipeline Name 3']]
```

```python
df = df.loc[(df['Plant Code'].isin(plant)) &
           (df['Sector Name'] == 'Electric Utility')]
df.tail(20)
```

```python
nanny = df.loc[(df['Natural Gas LDC Name'].isin([np.nan])) &
       (df['Natural Gas Pipeline Name 1'].isin([np.nan])) &
       (df['Natural Gas Pipeline Name 2'].isin([np.nan])) &
       (df['Natural Gas Pipeline Name 3'].isin([np.nan]))]
nanny.to_pickle('../pickleJar/NO_Pipeline.pkl')
```

```python
df = df.drop(nanny.index.values)
```
